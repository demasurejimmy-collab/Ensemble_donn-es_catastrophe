```python
df = pd.read_csv("global_disaster_response_2018_2024.csv", sep=",", encoding="utf-8")
print(df.columns)
```

```python
#convertir les types:
df["date"] = pd.to_datetime(df["date"], errors="coerce")
```

```python
# detecter des valeurs aberrante:
import numpy as np
q01 = df["economic_loss_usd"].quantile(0.01)
q99 = df["economic_loss_usd"].quantile(0.99)
df["economic_loss_usd"] =df["economic_loss_usd"].clip(q01, q99)

q01 =df["response_time_hours"].quantile(0.01)
q99 = df["response_time_hours"].quantile(0.99)
df["response_time_hours"] = df["response_time_hours"].clip(q01, q99)

df.describe()
```

```python
#creation fichier nettoyé

df.to_csv("catastrophe_donnees_nettoyer.csv" , index=False)
```
